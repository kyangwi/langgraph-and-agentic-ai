# 1. STRING OUTPUT PARSER

```python
template = """
Given the text beelow:

\n{text}
return a comma separated list. 
Do not use \n characters. Do not use new line. 
Do not use trailing double quotes. Do not return any description, just names.
"""

prompt = PromptTemplate.from_template(template)

chain = prompt | llm | StrOutputParser()

string = chain.invoke({'text':my_text})
print(string)
```

```python
chain = prompt | llm | CommaSeparatedListOutputParser()

list =chain.invoke({'text':my_text})
print(list)
```

```python
template = """
Given the text beelow:

\n{text}
Give me proper names onlyexcluding all comon one
use asterics and space for each name, 
do not use backsplash or quotes, asterics should be only at the begining 
of the name, remember to put space between asterics and name. 
Do not include any new line characters.
"""

prompt = PromptTemplate.from_template(template)

chain = prompt | llm | MarkdownListOutputParser()

result = chain.invoke({'text':my_text})
print(result)
```

# NUMBERED LIST

>*  NumberedListOutputParser is used to transform numbered list-like content into a Python list. Here’s an example:

```python
template = """
Give {number} names for a {domain} startup, return a numbered list. 
Do not wrap words into double quotes
"""

prompt = PromptTemplate.from_template(template)

chain = prompt | llm | NumberedListOutputParser()

result = chain.invoke({"number": 5, "domain": "tech"})

print("NUMBERED LIST", result)
```

# CUSTOM OUTPUT PARSERS

```python
from typing import List
from langchain_core.pydantic_v1 import BaseModel, Field

class StartupIdeas(BaseModel):
    """Startup ideas names"""
    names: List[str] = Field(..., description="Names for a startup")



template = "Give {number} names for a {domain} startup"

prompt = PromptTemplate.from_template(template)

structured_llm = llm.with_structured_output(StartupIdeas)

chain = prompt | structured_llm

result = chain.invoke({"number": 5, "domain": "tech"})

print("CUSTOM", result)
```
